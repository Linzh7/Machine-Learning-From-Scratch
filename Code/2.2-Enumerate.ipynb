```python
import DataReader
import numpy as np
from Cost import Cost
import matplotlib.pyplot as plt
```

```python
class Enumerate():
    def LinerRegressionEnumerate(data_x, data_y, maxRange=10, stepSize=0.1):
        cost = costSave = 0x7fffffff
        kSave = 0
        bSave = 0
        for k in np.arange(0, maxRange, stepSize):
            for b in np.arange(0, maxRange, stepSize):
                predict_y = np.array(data_x * k + b)
                cost = Cost.SquaredErrors(data_y, predict_y)
                if cost < costSave:
                    kSave = k
                    bSave = b
                    costSave = cost
                    #print(k, b, cost)
        return kSave, bSave, costSave
```

```python
data_x, data_y = DataReader.PrebuiltData.MyGivenData()
```

```python
print(data_x.shape==data_y.shape)
```

```python
import time
time_start=time.time()

k, b ,cost= Enumerate.LinerRegressionEnumerate(data_x, data_y, 3.1, 0.05)

time_end=time.time()

print(k,b,cost)
print('Time Cost: ',time_end-time_start)
```

```python
line_x = np.arange(1, 100)
line_y = line_x * k + b
```

```python
fig = plt.figure()
img = fig.add_subplot(111)
img.scatter(data_x, data_y, c='y')
img.scatter(line_x, line_y, marker = '.', c = 'b', s = 15)
plt.show()
```
