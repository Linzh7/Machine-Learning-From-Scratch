```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
k = 2
numberOfEachGroupPoint = 100
```

```python
dotMap = []
group = []
count = 0
for i in range(k):
    center = []
    center.append(np.random.randint(100))
    center.append(np.random.randint(100))
    group.append([])
    for j in range(numberOfEachGroupPoint):
        dot = []
        dot.append(center[0] + np.random.normal() * 8)
        dot.append(center[1] + np.random.normal() * 8)
        group[i].append(count)
        dotMap.append(dot)
        count += 1
dotMap = np.array(dotMap)
```

```python
data_X = dotMap[:, 0]
data_Y = dotMap[:, 1]
```

```python
fig = plt.figure()
img = fig.add_subplot(111)
cm = plt.cm.get_cmap('RdYlBu')
for i in range(k):
    img.scatter(data_X[group[i]], data_Y[group[i]], cmap=cm)
plt.show()
```
